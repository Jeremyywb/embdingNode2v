```python
import os
import pandas as pd

class FileTools(object):
    def __init__(self):
        self.dataroot = "../data/"
        self.cacheRoot = "../cache/"
        self.submitRoot = "../submit/"
        self.inDirDict = {"data":self.dataroot, "cache":self.cacheRoot, "sub":self.submitRoot }
    def useInfo(self):
        print("-------------------total info ------------------------")
        print(">----       use -> fileTool <- as tool class ---------")
        print("-------------------total info ------------------------")
        print("> ------------- @fun dir_initial  ------------- ")
        print(">--- use fileTool.dir_initial() ")
        print(">--- to initial model dir ----")
        print("> ------------- @fun dataHead  ------------- ")
        print(">--- use fileTool.dataHead(fname, headNum, inDirType) ")
        print(">--- @para:headNum choose to print out head of dataframe if 0 will print nothing  ")
        print(">--- @para: inDirType 1、 'data'  :read from dataRoot ")
        print(">--- @para: inDirType 2、 'cache' :read from cacheRoot ")
        print(">--- @return no return finilly del df read")
        print("> ------------- @fun readData  ------------- ")
        print(">--- use fileTool.readData(fname, headNum) ")
        print(">--- @func: read data from dataRoot ")
        print(">--- @para: fname, headNum as var name, headNum 0 print nothing ")
        print("> ------------- @fun readCache  ------------- ")
        print(">--- @func: same as  readData but from cacheRoot ")
        
    def mkdirInModel(self, dirRoot):
        if os.path.exists(self.dirRoot):
            print(self.dataroot,"already exists -----------")
        else:
            os.mkdir(dirRoot)       
    def dir_initial(self):
        self.mkdirInModel(self.dataroot)
        self.mkdirInModel(self.cacheRoot)
        self.mkdirInModel(self.submitRoot)
    def dataHead(self, fname, headNum, inDirType):
        df = pd.read_csv(self.inDirDict[inDirType] + fname )
        df.head(headNum)
        del df
    def readFile(self, fname, inDirType ):
        df =  pd.read_csv(self.inDirDict[inDirType] + fname )
        return df
    def readData(self, fname, headNum):
        df = self.readFile(fname,"data")
        if headNum != 0:
            df.head(headNum)
        return df
    def readCache(self, fname, headNum):
        df = self.readFile(fname,"cache")
        if headNum != 0:
            df.head(headNum)
        return df

fileTool =  FileTools()
fileTool.useInfo()
```
